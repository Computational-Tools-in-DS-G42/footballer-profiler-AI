## **Data Preprocessing**:


```python
# Removing unnacesarry the columns

print("Shape: " + str(init_df.shape))
cols_to_drop = ["club", "motm", "rating", "offsides_won", "league",  "fee", "traded", "club_position"]
df_clean = init_df.drop(columns=cols_to_drop)
print("Shape: " + str(df_clean.shape))
```

```python
#Filtering out players that played less than 90 mins
print("Before: " + str(df_clean["mins"].min()) + "mins")
df_clean = df_clean[df_clean["mins"] >= 90]
print("After: " + str(df_clean["mins"].min()) + "mins")
print("Shape: " + str(df_clean.shape))
```

```python
from unidecode import unidecode
from thefuzz import process
from datasketch import MinHash, MinHashLSH
```

```python
def clean_name(name):
    return unidecode(str(name).lower().strip())

df_clean['clean_name'] = df_clean['name'].apply(clean_name)
init_df2['clean_name'] = init_df2['name'].apply(clean_name)
```

```python
# Function to create a MinHash for a given string
def create_minhash(s):
    m = MinHash(num_perm=128)
    for token in set(s):  # you can also split by whitespace for more robustness
        m.update(token.encode('utf8'))
    return m

# Build the LSH index
lsh = MinHashLSH(threshold=0.5, num_perm=128)  # threshold ~ similarity cutoff
minhash_dict = {}

for idx, row in init_df2.iterrows():
    m = create_minhash(row['clean_name'])
    lsh.insert(str(idx), m)
    minhash_dict[str(idx)] = (m, row['clean_name'])
```

```python
enriched = []

for _, row in df_clean.iterrows():
    player = row['name']
    c_name = row['clean_name']
    m = create_minhash(c_name)

    # Find LSH candidates
    candidates = lsh.query(m)

    if candidates:
        # Compare only against a small set of likely matches
        best_match = None
        best_score = 0
        for cand in candidates:
            cand_name = minhash_dict[cand][1]
            score = fuzz.WRatio(c_name, cand_name)
            if score > best_score:
                best_match = cand_name
                best_score = score

        if best_score >= 90:
            matched_row = init_df2.loc[init_df2['clean_name'] == best_match]
            if not matched_row.empty:
                height = matched_row['height_in_cm'].values[0]
                foot = matched_row['foot'].values[0]
            else:
                height, foot = None, None
        else:
            best_match, best_score, height, foot = None, 0, None, None
    else:
        best_match, best_score, height, foot = None, 0, None, None

    enriched.append({
        'name': player,
        'matched_name': best_match,
        'match_score': best_score,
        'height': height,
        'preferred_foot': foot
    })

enriched_df = pd.DataFrame(enriched)
```
